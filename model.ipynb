```python
import pandas as pd
import os
import csv

# Fungsi deteksi delimiter otomatis
def detect_delimiter(file_path, sample_size=2048):
    with open(file_path, 'r', encoding='utf-8') as f:
        sample = f.read(sample_size)
        
        # Hitung kemunculan delimiter yang umum
        delimiters = [',', ';', '\t', '|']
        delimiter_counts = {}
        
        for delimiter in delimiters:
            delimiter_counts[delimiter] = sample.count(delimiter)
        
        # Pilih delimiter dengan count tertinggi
        best_delimiter = max(delimiter_counts, key=delimiter_counts.get)
        
        # Jika tidak ada delimiter yang ditemukan, gunakan sniffer
        if delimiter_counts[best_delimiter] == 0:
            sniffer = csv.Sniffer()
            try:
                dialect = sniffer.sniff(sample, delimiters=',;\t|')
                return dialect.delimiter
            except csv.Error:
                return ','  # fallback default
        
        return best_delimiter

# Daftar file CSV
file_list = [
    '001_reviews_jaksel_resto cina_10.csv',
    '002_reviews_jaksel_resto jepang_11.csv',
    '003_reviews_jaksel_resto western_8.csv',
    '004_reviews_jakut_resto jepang_11.csv',
    '005_reviews_jakut_resto china_8.csv',
    '006_reviews_jakut_resto indo_13.csv',
    '007_reviews_jaksel_fastfood_20.csv',
    '008_reviews_jaktim_resto khas indonesia_17.csv',
]

# Base directory tempat file berada dan hasil disimpan
base_dir = '/Users/tiarasabrina/Documents/PROJECT/Scrape google maps/fix_scrapped copy'

# Penampung data
all_reviews = []
resto_list = []
global_resto_counter = 1

# Mulai proses setiap file
for file in file_list:
    file_path = os.path.join(base_dir, file)
    delimiter = detect_delimiter(file_path)
    
    print(f"Processing {file} with delimiter: '{delimiter}'")
    
    try:
        df = pd.read_csv(file_path, encoding='utf-8', sep=delimiter)
    except Exception as e:
        print(f"Gagal membaca file {file}: {e}")
        # Coba dengan error handling yang lebih robust
        try:
            df = pd.read_csv(file_path, encoding='utf-8', sep=delimiter, 
                           on_bad_lines='skip')
            print(f"Berhasil membaca {file} dengan skip bad lines")
        except Exception as e2:
            # Coba dengan quoting yang berbeda
            try:
                df = pd.read_csv(file_path, encoding='utf-8', sep=delimiter, 
                               quoting=csv.QUOTE_ALL)
                print(f"Berhasil membaca {file} dengan quote all")
            except Exception as e3:
                print(f"Tetap gagal membaca file {file}: {e3}")
                continue

    if 'restaurant_id' not in df.columns:
        print(f"File {file} TIDAK punya kolom 'restaurant_id'. Kolom yang tersedia: {df.columns.tolist()}")
        continue

    df['restaurant_id'] = df['restaurant_id'].astype(str)
    
    # Dapatkan unique restaurant_id dalam urutan kemunculan pertama
    unique_ids = df['restaurant_id'].drop_duplicates().tolist()
    
    # Buat mapping dari original_id ke new_id yang berurutan global
    local_map = {}
    for orig_id in unique_ids:
        local_map[orig_id] = global_resto_counter
        
        # Simpan ke resto_list
        resto_list.append({
            'restaurant_id': global_resto_counter,
            'original_id': orig_id,
            'source_file': file
        })
        
        # Increment global counter
        global_resto_counter += 1

    # Apply mapping ke dataframe
    df['restaurant_id'] = df['restaurant_id'].map(local_map)
    
    print(f"File {file} - Processed {len(unique_ids)} unique restaurants")
    print(f"Original IDs: {unique_ids}")
    print(f"New IDs: {list(local_map.values())}")
    
    all_reviews.append(df)

# Gabung dan proses akhir
if all_reviews:
    review = pd.concat(all_reviews, ignore_index=True)
    review['review_id'] = range(1, len(review) + 1)
    review['user_id'] = range(1, len(review) + 1)

    review = review.dropna(subset=['review_text'])

    expected_columns = ['review_id', 'user_id', 'restaurant_id', 'review_text', 'rating', 'timestamp']
    review = review[[col for col in expected_columns if col in review.columns]]

    resto = pd.DataFrame(resto_list)[['restaurant_id', 'original_id', 'source_file']]

    # Simpan hasil
    review_path = os.path.join(base_dir, 'reviews.csv')
    resto_path = os.path.join(base_dir, 'restaurants.csv')

    review.to_csv(review_path, index=False, encoding='utf-8')
    resto.to_csv(resto_path, index=False, encoding='utf-8')

    print("\nReview dan restoran berhasil diproses dan disimpan:")
    print(f"- {review_path}")
    print(f"- {resto_path}")
    print(f"\nTotal restaurants processed: {len(resto_list)}")
    print(f"Total reviews processed: {len(review)}")
else:
    print("Tidak ada file yang berhasil diproses.")
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Core libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, haversine_distances
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import TruncatedSVD
import pickle
import re
import math

# Text processing
import nltk

# Download and verify NLTK data first
def setup_nltk():
    """Download and setup NLTK data"""
    try:
        nltk.data.find('tokenizers/punkt')
        print("NLTK punkt tokenizer already available")
    except LookupError:
        print("Downloading NLTK punkt tokenizer...")
        nltk.download('punkt', quiet=True)
    
    try:
        nltk.data.find('corpora/stopwords')
        print("NLTK stopwords already available")
    except LookupError:
        print("Downloading NLTK stopwords...")
        nltk.download('stopwords', quiet=True)

# Setup NLTK first
setup_nltk()

# Now import NLTK components
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# BERT
try:
    from sentence_transformers import SentenceTransformer
    BERT_AVAILABLE = True
except ImportError:
    print("Warning: sentence-transformers not available. BERT features will be disabled.")
    BERT_AVAILABLE = False

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

class RestaurantRecommendationSystem:
    def __init__(self):
        """Initialize the recommendation system"""
        self.tfidf_vectorizer = None
        self.tfidf_matrix = None
        self.bert_model = None
        self.bert_embeddings = None
        self.cosine_sim_tfidf = None
        self.cosine_sim_bert = None
        self.scaler = StandardScaler()
        self.price_scaler = MinMaxScaler()
        self.stemmer = PorterStemmer()
        
        # Initialize stopwords with error handling
        try:
            indonesian_stops = set(stopwords.words('indonesian'))
        except:
            print("Warning: Indonesian stopwords not available, using English only")
            indonesian_stops = set()
        
        try:
            english_stops = set(stopwords.words('english'))
        except:
            print("Warning: English stopwords not available, using minimal set")
            english_stops = {'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for', 'of', 'with', 'by'}
        
        self.stop_words = indonesian_stops.union(english_stops)
        
    def preprocess_text(self, text):
        """Clean and preprocess text data"""
        if pd.isna(text):
            return ""
        
        # Convert to lowercase
        text = str(text).lower()
        
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize with error handling
        try:
            tokens = word_tokenize(text)
        except:
            # Fallback to simple split if word_tokenize fails
            tokens = text.split()
        
        # Remove stopwords and stem
        processed_tokens = []
        for token in tokens:
            if token not in self.stop_words and len(token) > 2:
                try:
                    processed_tokens.append(self.stemmer.stem(token))
                except:
                    processed_tokens.append(token)
        
        return ' '.join(processed_tokens)
    
    def load_data(self, review_df, resto_df):
        """Load and prepare data"""
        print("Loading and preparing data...")
        
        self.review_df = review_df.copy()
        self.resto_df = resto_df.copy()
        
        # Handle missing values
        self.resto_df['description'] = self.resto_df['description'].fillna('')
        self.resto_df['category'] = self.resto_df['category'].fillna('Unknown')
        
        # Create combined text features
        self.resto_df['combined_features'] = (
            self.resto_df['name'].fillna('') + ' ' +
            self.resto_df['category'].fillna('') + ' ' +
            self.resto_df['description'].fillna('')
        )
        
        # Preprocess text
        print("Preprocessing text features...")
        self.resto_df['processed_features'] = self.resto_df['combined_features'].apply(
            self.preprocess_text
        )
        
        # Calculate average ratings
        avg_ratings = self.review_df.groupby('restaurant_id')['rating'].agg(['mean', 'count']).reset_index()
        avg_ratings.columns = ['id_resto', 'avg_rating', 'review_count']
        
        # Merge with restaurant data
        self.resto_df = self.resto_df.merge(avg_ratings, on='id_resto', how='left')
        self.resto_df['avg_rating'] = self.resto_df['avg_rating'].fillna(0)
        self.resto_df['review_count'] = self.resto_df['review_count'].fillna(0)
        
        # Calculate price range (if available)
        if 'harga_terendah' in self.resto_df.columns and 'harga_tertinggi' in self.resto_df.columns:
            self.resto_df['price_range'] = (
                self.resto_df['harga_terendah'].fillna(0) + 
                self.resto_df['harga_tertinggi'].fillna(0)
            ) / 2
            self.resto_df['price_range'] = self.resto_df['price_range'].fillna(0)
        else:
            self.resto_df['price_range'] = 0

        # Convert lat/lon to float in case they are strings
        self.resto_df['latitude'] = pd.to_numeric(self.resto_df['latitude'], errors='coerce')
        self.resto_df['longitude'] = pd.to_numeric(self.resto_df['longitude'], errors='coerce')

            
        print(f"Loaded {len(self.resto_df)} restaurants and {len(self.review_df)} reviews")
        
    def build_tfidf_model(self):
        """Build TF-IDF model for content-based filtering"""
        print("Building TF-IDF model...")
        
        # Create TF-IDF matrix
        self.tfidf_vectorizer = TfidfVectorizer(
            max_features=5000,
            ngram_range=(1, 2),
            min_df=1,
            max_df=0.8,
            stop_words=None 
        )
        
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(
            self.resto_df['processed_features']
        )
        
        # Calculate cosine similarity
        self.cosine_sim_tfidf = cosine_similarity(self.tfidf_matrix)
        
        print(f"TF-IDF matrix shape: {self.tfidf_matrix.shape}")
        
    def build_bert_model(self):
        """Build BERT model for semantic similarity"""
        if not BERT_AVAILABLE:
            print("BERT model not available. Skipping BERT embeddings.")
            return
            
        print("Loading BERT model and creating embeddings...")
        
        try:
            # Load pre-trained BERT model
            self.bert_model = SentenceTransformer('all-MiniLM-L6-v2')
            
            # Create embeddings
            texts = self.resto_df['combined_features'].fillna('').tolist()
            self.bert_embeddings = self.bert_model.encode(texts, show_progress_bar=True)
            
            # Calculate cosine similarity
            self.cosine_sim_bert = cosine_similarity(self.bert_embeddings)
            
            print(f"BERT embeddings shape: {self.bert_embeddings.shape}")
        except Exception as e:
            print(f"Error building BERT model: {e}")
            print("BERT recommendations will not be available.")
        
    def calculate_geo_distance(self, lat1, lon1, lat2, lon2):
        """Calculate distance between two points using Haversine formula"""
        if pd.isna(lat1) or pd.isna(lon1) or pd.isna(lat2) or pd.isna(lon2):
            return float('inf')
            
        # Convert to radians
        lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
        
        # Haversine formula
        dlat = lat2 - lat1
        dlon = lon2 - lon1
        a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
        c = 2 * math.asin(math.sqrt(a))
        r = 6371  # Radius of earth in kilometers
        
        return c * r
    
    def get_content_recommendations(self, restaurant_id, method='tfidf', top_n=10):
        """Get content-based recommendations"""
        try:
            idx = self.resto_df[self.resto_df['id_resto'] == restaurant_id].index[0]
        except IndexError:
            print(f"Restaurant ID {restaurant_id} not found")
            return pd.DataFrame()
            
        if method == 'tfidf':
            if self.cosine_sim_tfidf is None:
                print("TF-IDF model not built. Please run build_tfidf_model() first.")
                return pd.DataFrame()
            sim_scores = list(enumerate(self.cosine_sim_tfidf[idx]))
        elif method == 'bert':
            if self.cosine_sim_bert is None:
                print("BERT model not built or not available. Please run build_bert_model() first.")
                return pd.DataFrame()
            sim_scores = list(enumerate(self.cosine_sim_bert[idx]))
        else:
            raise ValueError("Method must be 'tfidf' or 'bert'")
            
        # Sort by similarity scores
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        
        # Get top N similar restaurants (excluding the input restaurant)
        sim_scores = sim_scores[1:top_n+1]
        
        # Get restaurant indices
        restaurant_indices = [i[0] for i in sim_scores]
        
        # Return restaurant details with similarity scores
        recommendations = self.resto_df.iloc[restaurant_indices].copy()
        recommendations['similarity_score'] = [score[1] for score in sim_scores]
        
        return recommendations
    
    def get_geo_filtered_recommendations(self, user_lat, user_lon, max_distance_km=5, 
                                      category_filter=None, min_rating=0, top_n=10):
        """Get recommendations based on geographic proximity and filters"""
        recommendations = []
        
        for idx, restaurant in self.resto_df.iterrows():
            distance = self.calculate_geo_distance(
                user_lat, user_lon, 
                restaurant['latitude'], restaurant['longitude']
            )
            
            # Apply filters
            if distance <= max_distance_km:
                if category_filter is None or restaurant['category'] == category_filter:
                    if restaurant['avg_rating'] >= min_rating:
                        rec = restaurant.copy()
                        rec['distance_km'] = distance
                        recommendations.append(rec)
        
        # Convert to DataFrame and sort by distance
        if recommendations:
            recommendations_df = pd.DataFrame(recommendations)
            recommendations_df = recommendations_df.sort_values('distance_km').head(top_n)
            return recommendations_df
        else:
            return pd.DataFrame()
    
    def get_hybrid_recommendations(self, restaurant_id=None, user_lat=None, user_lon=None,
                                 method='tfidf', max_distance_km=10, weight_content=0.7, 
                                 weight_geo=0.3, top_n=10):
        """Get hybrid recommendations combining content and geo filtering"""
        
        if restaurant_id is not None:
            # Content-based recommendations
            content_recs = self.get_content_recommendations(restaurant_id, method, top_n=50)
            
            if len(content_recs) == 0:
                print("No content-based recommendations found")
                return pd.DataFrame()
            
            if user_lat is not None and user_lon is not None:
                # Add distance information
                content_recs['distance_km'] = content_recs.apply(
                    lambda x: self.calculate_geo_distance(
                        user_lat, user_lon, x['latitude'], x['longitude']
                    ), axis=1
                )
                
                # Filter by distance
                content_recs = content_recs[content_recs['distance_km'] <= max_distance_km]
                
                # Calculate hybrid score
                # Normalize similarity scores and distance scores
                if len(content_recs) > 0:
                    max_sim = content_recs['similarity_score'].max()
                    min_dist = content_recs['distance_km'].min()
                    max_dist = content_recs['distance_km'].max()
                    
                    content_recs['norm_similarity'] = content_recs['similarity_score'] / max_sim if max_sim > 0 else 0
                    
                    if max_dist > min_dist:
                        content_recs['norm_distance'] = 1 - (content_recs['distance_km'] - min_dist) / (max_dist - min_dist)
                    else:
                        content_recs['norm_distance'] = 1
                    
                    content_recs['hybrid_score'] = (
                        weight_content * content_recs['norm_similarity'] + 
                        weight_geo * content_recs['norm_distance']
                    )
                    
                    content_recs = content_recs.sort_values('hybrid_score', ascending=False)
            
            return content_recs.head(top_n)
        
        elif user_lat is not None and user_lon is not None:
            # Pure geo-based recommendations
            return self.get_geo_filtered_recommendations(
                user_lat, user_lon, max_distance_km, top_n=top_n
            )
        
        else:
            print("Please provide either restaurant_id or user coordinates")
            return pd.DataFrame()
    
    def get_popular_recommendations(self, category=None, min_reviews=5, top_n=10):
        """Get popular restaurants based on ratings and review count"""
        filtered_df = self.resto_df.copy()
        
        if category:
            filtered_df = filtered_df[filtered_df['category'] == category]
        
        # Filter by minimum reviews
        filtered_df = filtered_df[filtered_df['review_count'] >= min_reviews]
        
        if len(filtered_df) == 0:
            print(f"No restaurants found with minimum {min_reviews} reviews")
            return pd.DataFrame()
        
        # Calculate popularity score (weighted rating)
        C = filtered_df['avg_rating'].mean()  # Mean rating across all restaurants
        m = filtered_df['review_count'].quantile(0.6)  # Minimum reviews required
        
        def weighted_rating(x):
            v = x['review_count']
            R = x['avg_rating']
            return (v/(v+m) * R) + (m/(m+v) * C)
        
        filtered_df['popularity_score'] = filtered_df.apply(weighted_rating, axis=1)
        
        return filtered_df.sort_values('popularity_score', ascending=False).head(top_n)
    
    def save_model(self, filepath):
        """Save the trained model"""
        model_data = {
            'tfidf_vectorizer': self.tfidf_vectorizer,
            'tfidf_matrix': self.tfidf_matrix,
            'bert_embeddings': self.bert_embeddings,
            'cosine_sim_tfidf': self.cosine_sim_tfidf,
            'cosine_sim_bert': self.cosine_sim_bert,
            'resto_df': self.resto_df,
            'review_df': self.review_df
        }
        
        with open(filepath, 'wb') as f:
            pickle.dump(model_data, f)
        print(f"Model saved to {filepath}")
    
    def load_model(self, filepath):
        """Load a trained model"""
        with open(filepath, 'rb') as f:
            model_data = pickle.load(f)
        
        self.tfidf_vectorizer = model_data['tfidf_vectorizer']
        self.tfidf_matrix = model_data['tfidf_matrix']
        self.bert_embeddings = model_data['bert_embeddings']
        self.cosine_sim_tfidf = model_data['cosine_sim_tfidf']
        self.cosine_sim_bert = model_data['cosine_sim_bert']
        self.resto_df = model_data['resto_df']
        self.review_df = model_data['review_df']
        
        print(f"Model loaded from {filepath}")
    
    def evaluate_model(self, test_cases=None):
        """Evaluate model performance"""
        print("\n=== MODEL EVALUATION ===")
        print(f"Total restaurants: {len(self.resto_df)}")
        print(f"Total reviews: {len(self.review_df)}")
        
        if self.tfidf_matrix is not None:
            print(f"TF-IDF matrix density: {self.tfidf_matrix.nnz / (self.tfidf_matrix.shape[0] * self.tfidf_matrix.shape[1]):.4f}")
            print(f"Average cosine similarity (TF-IDF): {np.mean(self.cosine_sim_tfidf):.4f}")
        
        if self.cosine_sim_bert is not None:
            print(f"Average cosine similarity (BERT): {np.mean(self.cosine_sim_bert):.4f}")
        
        # Category distribution
        print("\nCategory distribution:")
        print(self.resto_df['category'].value_counts().head())
        
        # Rating distribution
        print(f"\nAverage rating: {self.resto_df['avg_rating'].mean():.2f}")
        print(f"Rating std: {self.resto_df['avg_rating'].std():.2f}")

def main():
    review_df = review.copy()  # Use the cleaned review data from previous steps
    resto_df = resto.copy()  # Use the cleaned restaurant data from previous steps
    
    # Initialize and train the recommendation system
    print("\nInitializing recommendation system...")
    recommender = RestaurantRecommendationSystem()
    
    # Load data
    recommender.load_data(review_df, resto_df)
    
    # Build models
    recommender.build_tfidf_model()
    if BERT_AVAILABLE:
        recommender.build_bert_model()
    
    # Evaluate model
    recommender.evaluate_model()
    
    print("\n=== TESTING RECOMMENDATIONS ===")
    
    # Test content-based recommendations (TF-IDF)
    print("\n1. Content-based recommendations (TF-IDF) for restaurant 1:")
    tfidf_recs = recommender.get_content_recommendations(1, method='tfidf', top_n=2)
    if len(tfidf_recs) > 0:
        print(tfidf_recs[['name', 'category', 'similarity_score']])
    else:
        print("No TF-IDF recommendations found")
    
    # Test content-based recommendations (BERT) if available
    if BERT_AVAILABLE and recommender.cosine_sim_bert is not None:
        print("\n2. Content-based recommendations (BERT) for restaurant 1:")
        bert_recs = recommender.get_content_recommendations(1, method='bert', top_n=2)
        if len(bert_recs) > 0:
            print(bert_recs[['name', 'category', 'similarity_score']])
        else:
            print("No BERT recommendations found")
    
    # Test geo-filtered recommendations
    print("\n3. Geo-filtered recommendations near (-6.24, 106.79):")
    geo_recs = recommender.get_geo_filtered_recommendations(
        -6.24, 106.79, max_distance_km=2, top_n=3
    )
    if len(geo_recs) > 0:
        print(geo_recs[['name', 'category', 'distance_km']])
    else:
        print("No geo-filtered recommendations found")
    
    # Test hybrid recommendations
    print("\n4. Hybrid recommendations:")
    hybrid_recs = recommender.get_hybrid_recommendations(
        restaurant_id=1, user_lat=-6.24, user_lon=106.79, 
        method='tfidf', max_distance_km=5, top_n=2
    )
    if len(hybrid_recs) > 0:
        print(hybrid_recs[['name', 'category', 'similarity_score', 'distance_km']])
    else:
        print("No hybrid recommendations found")
    
    # Test popular recommendations
    print("\n5. Popular recommendations:")
    popular_recs = recommender.get_popular_recommendations(min_reviews=1, top_n=3)
    if len(popular_recs) > 0:
        print(popular_recs[['name', 'category', 'avg_rating', 'popularity_score']])
    else:
        print("No popular recommendations found")
    
    # Save model
    print("\nSaving model...")
    try:
        recommender.save_model('restaurant_recommender.pkl')
    except Exception as e:
        print(f"Error saving model: {e}")
    
    print("\n=== RECOMMENDATION SYSTEM READY ===")
    return recommender

if __name__ == "__main__":
    recommender = main()
```

```python
# Practical usage script for your restaurant recommendation system
import pandas as pd
import numpy as np

def load_your_data():
    """Load your actual data - replace with your file paths"""
    
    review_df = pd.read_csv('reviews.csv')
    resto_df = pd.read_csv('restaurants.csv')
    
    # Rename columns to match the expected format
    if 'restaurant_id' in review_df.columns:
        review_df = review_df.rename(columns={'restaurant_id': 'restaurant_id'})
    
    if 'original_id' in resto_df.columns:
        resto_df = resto_df.rename(columns={'restaurant_id': 'id_resto'})
    
    print(f"Loaded {len(review_df)} reviews and {len(resto_df)} restaurants")
    print("Review columns:", review_df.columns.tolist())
    print("Restaurant columns:", resto_df.columns.tolist())
    
    return review_df, resto_df

def train_and_save_model():
    """Train the model with your data and save it"""
    
    print("=== TRAINING RESTAURANT RECOMMENDATION MODEL ===")
    
    # Load data
    review_df, resto_df = load_your_data()
    
    # Initialize the recommender
    recommender = RestaurantRecommendationSystem()
    
    # Load and prepare data
    print("\nPreparing data...")
    recommender.load_data(review_df, resto_df)
    
    # Build TF-IDF model
    print("\nBuilding TF-IDF model...")
    recommender.build_tfidf_model()
    
    # Build BERT model
    print("\nBuilding BERT model...")
    recommender.build_bert_model()
    
    # Evaluate the model
    print("\nEvaluating model...")
    recommender.evaluate_model()
    
    # Save the trained model
    print("\nSaving model...")
    recommender.save_model('restaurant_recommender.pkl')
    
    print("\nModel training completed and saved!")
    return recommender

def use_trained_model():
    """Load and use the trained model for recommendations"""
    
    print("=== USING TRAINED MODEL FOR RECOMMENDATIONS ===")
    
    # Initialize and load the trained model
    recommender = RestaurantRecommendationSystem()
    recommender.load_model('restaurant_recommender.pkl')
    
    # Example 1: Content-based recommendations using BERT
    print("\n1. Content-based recommendations (BERT):")
    restaurant_id = 1  # Change this to any restaurant ID from your data
    
    try:
        bert_recommendations = recommender.get_content_recommendations(
            restaurant_id=restaurant_id, 
            method='bert', 
            top_n=5
        )
        
        if len(bert_recommendations) > 0:
            print(f"\nRestaurants similar to ID {restaurant_id}:")
            for idx, row in bert_recommendations.iterrows():
                print(f"- {row['name']} ({row['category']}) - Similarity: {row['similarity_score']:.3f}")
        else:
            print("No recommendations found")
            
    except Exception as e:
        print(f"Error getting content recommendations: {e}")
    
    # Example 2: Geo-filtered recommendations
    print("\n2. Geo-filtered recommendations:")
    user_lat = -6.2391  # Your latitude
    user_lon = 106.7906  # Your longitude
    
    geo_recommendations = recommender.get_geo_filtered_recommendations(
        user_lat=user_lat,
        user_lon=user_lon,
        max_distance_km=5,  # Within 5km
        min_rating=3.0,     # Minimum rating 3.0
        top_n=5
    )
    
    if len(geo_recommendations) > 0:
        print(f"\nRestaurants within 5km of ({user_lat}, {user_lon}):")
        for idx, row in geo_recommendations.iterrows():
            print(f"- {row['name']} ({row['category']}) - Distance: {row['distance_km']:.2f}km, Rating: {row['avg_rating']:.1f}")
    else:
        print("No restaurants found within the specified distance")
    
    # Example 3: Hybrid recommendations
    print("\n3. Hybrid recommendations (Content + Geo):")
    
    try:
        hybrid_recommendations = recommender.get_hybrid_recommendations(
            restaurant_id=restaurant_id,
            user_lat=user_lat,
            user_lon=user_lon,
            method='bert',
            max_distance_km=10,
            weight_content=0.7,  # 70% content similarity
            weight_geo=0.3,      # 30% geographic proximity
            top_n=5
        )
        
        if len(hybrid_recommendations) > 0:
            print(f"\nHybrid recommendations for restaurant {restaurant_id}:")
            for idx, row in hybrid_recommendations.iterrows():
                print(f"- {row['name']} ({row['category']}) - Score: {row['hybrid_score']:.3f}, Distance: {row['distance_km']:.2f}km")
        else:
            print("No hybrid recommendations found")
            
    except Exception as e:
        print(f"Error getting hybrid recommendations: {e}")
    
    # Example 4: Popular restaurants
    print("\n4. Popular restaurants:")
    
    popular_recommendations = recommender.get_popular_recommendations(
        category=None,      # All categories, or specify like 'Restoran China'
        min_reviews=2,      # Minimum 2 reviews
        top_n=5
    )
    
    if len(popular_recommendations) > 0:
        print("\nMost popular restaurants:")
        for idx, row in popular_recommendations.iterrows():
            print(f"- {row['name']} ({row['category']}) - Rating: {row['avg_rating']:.1f} ({int(row['review_count'])} reviews)")
    else:
        print("No popular restaurants found")
    
    # Example 5: Category-specific recommendations
    print("\n5. Category-specific popular restaurants:")
    
    categories = recommender.resto_df['category'].unique()[:3]  # Get first 3 categories
    
    for category in categories:
        if pd.notna(category) and category != 'Unknown':
            category_recs = recommender.get_popular_recommendations(
                category=category,
                min_reviews=1,
                top_n=3
            )
            
            if len(category_recs) > 0:
                print(f"\n{category}:")
                for idx, row in category_recs.iterrows():
                    print(f"  - {row['name']} - Rating: {row['avg_rating']:.1f}")

def interactive_recommendation():
    """Interactive recommendation function"""
    
    print("=== INTERACTIVE RESTAURANT RECOMMENDATION ===")
    
    # Load the trained model
    recommender = RestaurantRecommendationSystem()
    
    try:
        recommender.load_model('restaurant_recommender.pkl')
    except FileNotFoundError:
        print("Trained model not found. Please run train_and_save_model() first.")
        return
    
    print("\nAvailable restaurants:")
    restaurants = recommender.resto_df[['id_resto', 'name', 'category']].head(10)
    print(restaurants.to_string(index=False))
    
    while True:
        print("\n" + "="*50)
        print("Choose recommendation type:")
        print("1. Content-based (similar restaurants)")
        print("2. Geo-filtered (nearby restaurants)")
        print("3. Hybrid (content + location)")
        print("4. Popular restaurants")
        print("5. Exit")
        
        choice = input("\nEnter your choice (1-5): ").strip()
        
        if choice == '1':
            restaurant_id = input("Enter restaurant ID: ").strip()
            try:
                restaurant_id = int(restaurant_id)
                method = input("Choose method (tfidf/bert) [bert]: ").strip() or 'bert'
                
                recs = recommender.get_content_recommendations(restaurant_id, method, top_n=5)
                
                if len(recs) > 0:
                    print(f"\nSimilar restaurants to ID {restaurant_id}:")
                    for idx, row in recs.iterrows():
                        print(f"- {row['name']} ({row['category']}) - Similarity: {row['similarity_score']:.3f}")
                else:
                    print("No recommendations found")
                    
            except ValueError:
                print("Please enter a valid restaurant ID")
            except Exception as e:
                print(f"Error: {e}")
        
        elif choice == '2':
            try:
                lat = float(input("Enter your latitude: ").strip())
                lon = float(input("Enter your longitude: ").strip())
                max_dist = float(input("Max distance in km [5]: ").strip() or 5)
                
                recs = recommender.get_geo_filtered_recommendations(lat, lon, max_dist, top_n=5)
                
                if len(recs) > 0:
                    print(f"\nRestaurants within {max_dist}km:")
                    for idx, row in recs.iterrows():
                        print(f"- {row['name']} ({row['category']}) - Distance: {row['distance_km']:.2f}km, Rating: {row['avg_rating']:.1f}")
                else:
                    print("No restaurants found within the specified distance")
                    
            except ValueError:
                print("Please enter valid coordinates")
            except Exception as e:
                print(f"Error: {e}")
        
        elif choice == '3':
            try:
                restaurant_id = int(input("Enter restaurant ID: ").strip())
                lat = float(input("Enter your latitude: ").strip())
                lon = float(input("Enter your longitude: ").strip())
                max_dist = float(input("Max distance in km [10]: ").strip() or 10)
                method = input("Choose method (tfidf/bert) [bert]: ").strip() or 'bert'
                
                recs = recommender.get_hybrid_recommendations(
                    restaurant_id=restaurant_id,
                    user_lat=lat,
                    user_lon=lon,
                    method=method,
                    max_distance_km=max_dist,
                    top_n=5
                )
                
                if len(recs) > 0:
                    print(f"\nHybrid recommendations for restaurant {restaurant_id}:")
                    for idx, row in recs.iterrows():
                        print(f"- {row['name']} ({row['category']}) - Score: {row['hybrid_score']:.3f}, Distance: {row['distance_km']:.2f}km")
                else:
                    print("No hybrid recommendations found")
                    
            except ValueError:
                print("Please enter valid inputs")
            except Exception as e:
                print(f"Error: {e}")
        
        elif choice == '4':
            try:
                category = input("Enter category (optional, press Enter for all): ").strip() or None
                min_reviews = int(input("Minimum reviews [2]: ").strip() or 2)
                
                recs = recommender.get_popular_recommendations(
                    category=category,
                    min_reviews=min_reviews,
                    top_n=5
                )
                
                if len(recs) > 0:
                    print(f"\nPopular restaurants{' in ' + category if category else ''}:")
                    for idx, row in recs.iterrows():
                        print(f"- {row['name']} ({row['category']}) - Rating: {row['avg_rating']:.1f} ({int(row['review_count'])} reviews)")
                else:
                    print("No popular restaurants found with the specified criteria")
                    
            except ValueError:
                print("Please enter valid numbers")
            except Exception as e:
                print(f"Error: {e}")
        
        elif choice == '5':
            print("Goodbye!")
            break
        
        else:
            print("Invalid choice. Please enter 1-5.")

def analyze_recommendations():
    """Analyze recommendation quality and system performance"""
    
    print("=== RECOMMENDATION SYSTEM ANALYSIS ===")
    
    # Load the trained model
    recommender = RestaurantRecommendationSystem()
    
    try:
        recommender.load_model('restaurant_recommender.pkl')
    except FileNotFoundError:
        print("Trained model not found. Please run train_and_save_model() first.")
        return
    
    # Basic statistics
    print("\n📊 Dataset Statistics:")
    print(f"Total restaurants: {len(recommender.resto_df)}")
    print(f"Total reviews: {len(recommender.review_df)}")
    print(f"Average rating: {recommender.resto_df['avg_rating'].mean():.2f}")
    print(f"Restaurants with reviews: {len(recommender.resto_df[recommender.resto_df['review_count'] > 0])}")
    
    # Category distribution
    print("\nCategory Distribution:")
    category_counts = recommender.resto_df['category'].value_counts()
    for category, count in category_counts.head(10).items():
        print(f"  {category}: {count} restaurants")
    
    # Rating distribution
    print("\nRating Distribution:")
    rating_bins = pd.cut(recommender.resto_df['avg_rating'], bins=[0, 2, 3, 4, 5], labels=['Poor (0-2)', 'Fair (2-3)', 'Good (3-4)', 'Excellent (4-5)'])
    rating_dist = rating_bins.value_counts()
    for rating, count in rating_dist.items():
        print(f"  {rating}: {count} restaurants")
    
    # Geographic distribution
    print("\nGeographic Coverage:")
    valid_coords = recommender.resto_df.dropna(subset=['latitude', 'longitude'])
    if len(valid_coords) > 0:
        print(f"Restaurants with coordinates: {len(valid_coords)}")
        print(f"Latitude range: {valid_coords['latitude'].min():.4f} to {valid_coords['latitude'].max():.4f}")
        print(f"Longitude range: {valid_coords['longitude'].min():.4f} to {valid_coords['longitude'].max():.4f}")
    else:
        print("No restaurants with valid coordinates found")
    
    # Recommendation diversity analysis
    print("\nRecommendation Diversity Analysis:")
    test_restaurants = recommender.resto_df['id_resto'].head(5).tolist()
    
    all_recommended = set()
    for rest_id in test_restaurants:
        try:
            recs = recommender.get_content_recommendations(rest_id, method='tfidf', top_n=5)
            if len(recs) > 0:
                all_recommended.update(recs['id_resto'].tolist())
        except:
            pass
    
    diversity_score = len(all_recommended) / (len(test_restaurants) * 5) if test_restaurants else 0
    print(f"Diversity score: {diversity_score:.2f} (higher is more diverse)")
    
    # Coverage analysis
    coverage_score = len(all_recommended) / len(recommender.resto_df)
    print(f"Coverage score: {coverage_score:.2f} (fraction of restaurants that can be recommended)")

def batch_recommendations():
    """Generate batch recommendations for multiple scenarios"""
    
    print("=== BATCH RECOMMENDATION GENERATION ===")
    
    # Load the trained model
    recommender = RestaurantRecommendationSystem()
    
    try:
        recommender.load_model('restaurant_recommender.pkl')
    except FileNotFoundError:
        print("Trained model not found. Please run train_and_save_model() first.")
        return
    
    # Define multiple user scenarios
    scenarios = [
        {
            'name': 'Downtown User',
            'lat': -6.2088,
            'lon': 106.8456,
            'preferences': {'category': 'Restoran China', 'max_distance': 3}
        },
        {
            'name': 'South Jakarta User',
            'lat': -6.2615,
            'lon': 106.7810,
            'preferences': {'category': None, 'max_distance': 5}
        },
        {
            'name': 'Budget-conscious User',
            'lat': -6.2391,
            'lon': 106.7906,
            'preferences': {'max_price': 50000, 'max_distance': 7}
        }
    ]
    
    results = {}
    
    for scenario in scenarios:
        print(f"\nGenerating recommendations for {scenario['name']}:")
        
        # Geo-filtered recommendations
        geo_recs = recommender.get_geo_filtered_recommendations(
            user_lat=scenario['lat'],
            user_lon=scenario['lon'],
            max_distance_km=scenario['preferences']['max_distance'],
            category_filter=scenario['preferences'].get('category'),
            top_n=5
        )
        
        # Popular recommendations in area
        popular_recs = recommender.get_popular_recommendations(
            category=scenario['preferences'].get('category'),
            min_reviews=1,
            top_n=5
        )
        
        results[scenario['name']] = {
            'geo_recommendations': geo_recs,
            'popular_recommendations': popular_recs
        }
        
        print(f"  Found {len(geo_recs)} nearby restaurants")
        print(f"  Found {len(popular_recs)} popular restaurants")
        
        if len(geo_recs) > 0:
            print("  Top nearby restaurants:")
            for idx, row in geo_recs.head(3).iterrows():
                print(f"    - {row['name']} ({row['distance_km']:.1f}km)")
    
    # Save results
    print("\nSaving batch results...")
    for scenario_name, recs in results.items():
        filename = f"recommendations_{scenario_name.lower().replace(' ', '_')}.csv"
        if len(recs['geo_recommendations']) > 0:
            recs['geo_recommendations'].to_csv(filename, index=False)
            print(f"  Saved {filename}")
    
    return results

def main():
    """Main function to run the recommendation system"""
    
    print("RESTAURANT RECOMMENDATION SYSTEM")
    print("="*50)
    
    while True:
        print("\nChoose an option:")
        print("1. Train and save model")
        print("2. Use trained model")
        print("3. Interactive recommendations")
        print("4. Analyze system performance")
        print("5. Generate batch recommendations")
        print("6. Exit")
        
        choice = input("\nEnter your choice (1-6): ").strip()
        
        if choice == '1':
            try:
                train_and_save_model()
            except Exception as e:
                print(f"Error during training: {e}")
                print("Make sure your CSV files exist and have the correct columns.")
        
        elif choice == '2':
            try:
                use_trained_model()
            except Exception as e:
                print(f"Error using model: {e}")
        
        elif choice == '3':
            interactive_recommendation()
        
        elif choice == '4':
            try:
                analyze_recommendations()
            except Exception as e:
                print(f"Error during analysis: {e}")
        
        elif choice == '5':
            try:
                batch_recommendations()
            except Exception as e:
                print(f"Error generating batch recommendations: {e}")
        
        elif choice == '6':
            print("Thank you for using the Restaurant Recommendation System!")
            break
        
        else:
            print("Invalid choice. Please enter 1-6.")

if __name__ == "__main__":
    main()
```
